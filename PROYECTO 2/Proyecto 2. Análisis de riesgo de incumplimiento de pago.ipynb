Primero voy a dividirlos por grupo de **edad** y calcular la media y la mediana, despues analizare los mismos valores estadisticos al agrupar los clientes por **nivel educativo**.

```python
# Función para calcular la categoría de edad por decadas:
def age_group_dec(row):
    age=row['dob_years']
    
    if 18 <= age <= 30: 
        return '18-30'  
    if 30 < age <= 40:
        return '31-40'
    if 40 < age <= 50:
        return '41-50'
    if 50 < age <= 60:
        return '51-60'
    if 60 < age <= 70:
        return '61-70' 
    return '+70'      
```

```python
# Crear una nueva columna basada en la función
credit_scoring['age_group_dec'] = credit_scoring.apply(age_group_dec, axis=1)
```
